```python
import pandas as pd 
import scipy
import matplotlib.pyplot as plt 

df = pd.read_csv('data.csv', sep=',')
```

Resistance - сопротивление 50 резисторов одинакового номинала 

Time - время, за которое человек отсчитал 5 секунд в уме

```python
df['Resistance'].plot.kde()
plt.show()
```

```python
df['Time'].plot.kde()
plt.show()
```

Гипотеза: данные распределения близки к нормальному.

```python
from scipy import stats

resistance = df['Resistance']
time = df['Time']

print(stats.kstest('norm', 'norm', N=500))
print()

print(stats.kstest(resistance, 'norm', (resistance.mean(), resistance.std()), N=len(resistance)))
print(stats.kstest(time, 'norm', (time.mean(), time.std()), N=len(time)))
```

В итоге в двух случаях К-С-статистика мала, а P-значение велико, поэтому гипотезу отвергнуть нельзя. Кроме того, в первом случае значение statisic меньше чем во втором, а pvalue, наоборот, больше. Таким образом, распределение сопротивлений резисторов больше похоже на нормальное.
