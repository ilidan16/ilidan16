```python
import pandas as pd 
import scipy
import matplotlib.pyplot as plt 

df = pd.read_csv('data.csv')

data = pd.DataFrame(data={
    'Resistance': df['Resistance']
    })
data.plot.kde()
plt.show()
```

```python
from scipy import stats

print(stats.kstest(data, 'norm', (data.mean(), data.std()), N=len(data)))
```
